```python
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
sns.set_style(style='whitegrid')
```

```python
def q(T,h,epsilon):
    G = 500
    alpha = 0.8
    g = G*alpha
    d = 25 *10**(-3)
    T_s = 30 +273.15
    sigma = 5.67*10**(-8)
    T = float(T)
    q = d*(np.pi*(sigma*epsilon*(T_s**4 - T**4)+h*(T_s-T))-g)    
    return q
```

```python
# A)
epsilon = 0.95
h = 40
def raiz(epsilon,h):
    erro = 0.00001
    n = 100000
    a = 273.15
    b = 500
    u = q(a,h,epsilon)
    v = q(b,h,epsilon)
    d = u*v
    if d > 0:
        print('estime novo intervalo')
    else:
        for k in range (1,n):
            c = a + (b-a)/2
            w = q(c,h,epsilon)
            if np.abs(w)<= erro:
                
                T = c
                break
            d = w*u
            if d < 0:
                b = c
                v = w 
            else:
                a = c
                u = w    
        return T
T = raiz(epsilon,h)
print(f'T-Kelvin = {T:.2f} \nT-Celsius = {T-273.15:.2f}')
    
```

```python
#B)
h = 10
T_list = []
h_list = []
for i in range(0,10):
    T_i = raiz(epsilon,h)
    T_list.append(T_i)
    h_list.append(h)
    h+=19
plt.plot(h_list,T_list,color = 'green')
plt.plot(h_list,T_list,'o',color = 'red')
plt.xlabel('h [W/(m^2*K)]')
plt.ylabel('T [K]')
plt.title('Temperatura do ar X Coeficiente de convecção')
print(h_list)
print(T_list)
```

```python
epsilon = 0.095
h = 40
T_list=[]
epsilon_list =[]
for i in range(0,10):
    T_i = raiz(epsilon,h)
    T_list.append(T_i)
    epsilon_list.append(epsilon)
    epsilon += 0.095
plt.plot(epsilon_list,T_list,color = 'green')
plt.plot(epsilon_list,T_list,'o',color = 'red')
plt.xlabel('Emissividade')
plt.ylabel('T [K]')
plt.title('Temperatura do ar X Emissividade')
```
